# Pandas visualization

```python
import numpy as np
import pandas as pd
import seaborn as sns
sns.set()
%matplotlib inline
```

```python
df1 = pd.read_csv('df1',index_col=0)
df1.head()
```

```python
df2=pd.read_csv('df2')
df2.head()
```

## Visualising a dataframe directly

```python
df1['A'].hist(bins=20)
```

```python
df1['A'].plot(kind='hist')
```

```python
df1['A'].plot.hist()
```

```python
df2.plot.area(alpha=0.4)
```

```python
df2.plot.bar()
```

```python
df1['A'].plot.hist(bins=50)
```

```python
df1.reset_index().plot.line(x='index',y='B', figsize=(12,3),lw=1)
```

```python
df1.plot.scatter(x='A',y='B',c='C',cmap='coolwarm')
```

```python
df1.plot.scatter(x='A',y='B',s=df1['C']*10)
```

```python
df2.plot.box()
```

```python
df = pd.DataFrame(np.random.randn(1000,2),columns=['a','b'])
```

```python
df.plot.hexbin(x='a',y='b',gridsize=25,cmap='coolwarm')
```

```python
df2.plot.kde()
```
