# Matplotlib Basics

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = x**2
```

```python
x
```

```python
y
```

```python
# FUNCTIONAL
plt.plot(x,y,)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

## Subplot using functions

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# Object Oriented
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x, y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
```

## Subplots using OO

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

#axes.plot(x,y)
#plt.tight_layout()
#for current_ax in axes:
#    current_ax.plot(x,y)

axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')
```

## Figure Size and DPI

```python
fig, axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))

axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
fig.savefig('my_picture.png', dpi=200)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label='X^2')
ax.plot(x, x**3, label='X^3')

ax.legend(loc=0)
```

## Setting colors, line-widths, line-type

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, color='purple', lw=2, ls='--')  #Also RGB hex codes

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
