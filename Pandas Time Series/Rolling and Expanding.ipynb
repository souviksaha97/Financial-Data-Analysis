```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('time_data/walmart_stock.csv',
                index_col='Date',
                parse_dates=True)
```

```python
df.head()
```

```python
df['Open'].plot(figsize=(16,6))
```

## Rolling or Moving Average

```python
df.rolling(7).mean().head(14)
```

```python
df['Open'].plot()
df.rolling(window=7).mean()['Open'].plot(figsize=(16,6))
```

```python
df['Open'].plot()
df.rolling(window=30).mean()['Open'].plot(figsize=(16,6))
```

```python
df['Open'].plot()
df.rolling(window=90).mean()['Open'].plot(figsize=(16,6))
```

```python
df['Close 30 Day MA'] = df['Close'].rolling(window=30).mean()
df[['Close 30 Day MA', 'Close']].plot(figsize=(16,6))
```

```python
df['Close'].expanding().mean().plot(figsize=(16,6))
```

 ## Bollinger Bands

```python
# Close 20 MA
df['Close: 20 Day Mean'] = df['Close'].rolling(20).mean()

# Upper = 20MA + 2*std(20)
df['Upper'] = df['Close: 20 Day Mean'] + 2*(df['Close'].rolling(20).std())

# Lower = 20MA - 2*std(20)
df['Lower'] = df['Close: 20 Day Mean'] - 2*(df['Close'].rolling(20).std())

# Close
df[['Close', 'Close: 20 Day Mean', 'Upper', 'Lower']].tail(200).plot(figsize=(16,6))
```
